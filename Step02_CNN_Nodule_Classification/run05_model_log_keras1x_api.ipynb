```python
%pylab inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fcsv='./model_cnn_ct3d.h5-log.csv'
data=pd.read_csv(fcsv, sep=',')
#
plt.figure(figsize=(14,8))
plt.subplot(2,1,1)
plt.plot(data['loss'])
plt.plot(data['val_loss'])
plt.legend(['trn_loss', 'val_loss'])
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(data['acc'])
plt.plot(data['val_acc'])
plt.legend(['trn_acc', 'val_acc'])
plt.grid(True)
```
